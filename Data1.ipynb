## Splitting the dataset

Into training, test and validation data

```python
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
from sklearn import utils

# Columns place_final, points_final, points_tele_final and points_jury_final are components of place_final, so they're not useful and will be dropped
data_before_split = pd.read_csv("dataset_with_audio_features.csv").drop(columns=["place_final", "points_final", "points_tele_final", "points_jury_final"])

max_nr_of_contestants = data_before_split["place_contest"].max()
# Transforming place_contest to a uniform scale, i.e place divided by nr_of_contestants_in_that_year (times 43), to account for differences in the number of participating countries each year.
#beginning = 0
#end = 0
#while (end < data_before_split.shape[0]):
#    year = data_before_split.iloc[beginning,0]
#    while ( end < data_before_split.shape[0] and data_before_split.iloc[end,0] == year):
#        end += 1
#    nr_of_contestants = data_before_split.iloc[end-1,5]
#    for i in range(beginning, end):
#        data_before_split.iloc[i, 5] = round((data_before_split.iloc[i, 5]/nr_of_contestants)*max_nr_of_contestants,0)
#    beginning = end

# Columns that would provide too many dummies, or are not deemed useful, are also deleted. This includes year, the country id, performer, song, composers, lyricists, lyrics and youtube url
data_before_split = data_before_split.drop(columns=["year", "to_country_id", "performer", "song", "composers", "lyricists", "lyrics", "youtube_url"])

# Change all categorical columns into binary features using one-hot-encoding
data_before_split = pd.get_dummies(data_before_split, columns=["tonal.chords_key", "tonal.chords_scale", "tonal.key_key", "tonal.key_scale", "to_country"])

# Changing NaN values to 0 in the running order column
data_before_split["running_final"] = data_before_split["running_final"].fillna(0)

labels = data_before_split.place_contest
 
data_without_labels = data_before_split.drop(columns=["place_contest"])

# Splitting data into training and test-validation data
X_train, X_test_val, y_train, y_test_val = train_test_split(data_without_labels, labels, test_size=0.3, random_state=42)

# Splitting test-validation data into test and validation data
X_test, X_val, y_test, y_val = train_test_split(X_test_val, y_test_val, test_size=0.5, random_state=42)
```

## Calculating the accuracy of the models

Approaches are sourced from the course materials and https://towardsdatascience.com/how-to-evaluate-your-machine-learning-models-with-python-code-5f8d2d8d945b

1. Classification models

1.1 Random Forest Classifier

```python
from sklearn.metrics import accuracy_score

# Calculating the accuracy score, i.e correctl_predictions divided by total predictions
y_pred_rf = rf.predict(X_test)
print(y_pred_rf[:10])
print(y_test[:10])
acc_rf = accuracy_score(y_test, y_pred_rf)
print("Accuracy score: " + str(acc_rf))

# Calculating the accuracy score, but this time, a prediction is considered accurate if it is +-3 positions away from the true value
y_test_list = y_test.tolist()
acc_preds_rf = 0
for i in range(len(y_pred_rf)):
    if ((y_pred_rf[i]-3 <= y_test_list[i]) and (y_test_list[i] <= y_pred_rf[i] + 3)):
        acc_preds_rf += 1
        
acc_rf_range = acc_preds_rf/len(y_test_list)
print("Range accuracy: " + str(acc_rf_range))
```

1.2 KNN

```python
# Calculating the accuracy score, i.e correct_predictions divided by total predictions
y_pred_neigh = neigh.predict(X_test)
print(y_pred_neigh[:10])
print(y_test[:10])
acc_neigh = accuracy_score(y_test, y_pred_neigh)
print("Accuracy score: " + str(acc_neigh))

# Calculating the accuracy score, but this time, a prediction is considered accurate if it is +-3 positions away from the true value
y_test_list = y_test.tolist()
acc_preds_neigh = 0
for i in range(len(y_pred_neigh)):
    if ((y_pred_neigh[i]-3 <= y_test_list[i]) and (y_test_list[i] <= y_pred_neigh[i] + 3)):
        acc_preds_neigh += 1
        
acc_neigh_range = acc_preds_neigh/len(y_test_list)
print("Range accuracy: " + str(acc_neigh_range))
```

```python
print("Accuracy scores:")
print("RF: " + str(acc_rf), "KNN: " + str(acc_neigh))
print()
print("Range accuracies: ")
print("RF: " + str(acc_rf_range), "KNN: " + str(acc_neigh_range))
```

From this, we can see that the best classification model is the Random Forest Classifier.

2. Regression models

2.1 Linear regression

```python
from sklearn.metrics import r2_score

# Calculating the r-squared score of the model
y_pred_lr = lr.predict(X_test)
print(y_pred_lr[:10])
print(y_test[:10])
r2_score_lr = r2_score(y_test, y_pred_lr)
print("R-squared score: " + str(r2_score_lr))

# Calculating the range accuracy score of the model. Here, a prediction is considered accurate if it is +-3 positions away from the true value
y_test_list = y_test.tolist()
acc_preds_lr = 0
for i in range(len(y_pred_lr)):
    if ((y_pred_lr[i]-3 <= y_test_list[i]) and (y_test_list[i] <= y_pred_lr[i] + 3)):
        acc_preds_lr += 1
        
acc_lr_range = acc_preds_lr/len(y_test_list)
print("Range accuracy: " + str(acc_lr_range))
```

2.2 Ridge regression

```python
# Calculating the r-squared score of the model
y_pred_ridge = ridge.predict(X_test)
print(y_pred_ridge[:10])
print(y_test[:10])
r2_score_ridge = r2_score(y_test, y_pred_ridge)
print("R-squared score: " + str(r2_score_ridge))

# Calculating the range accuracy score of the model. Here, a prediction is considered accurate if it is +-3 positions away from the true value
y_test_list = y_test.tolist()
acc_preds_ridge = 0
for i in range(len(y_pred_ridge)):
    if ((y_pred_ridge[i]-3 <= y_test_list[i]) and (y_test_list[i] <= y_pred_ridge[i] + 3)):
        acc_preds_ridge += 1
        
acc_ridge_range = acc_preds_ridge/len(y_test_list)
print("Range accuracy: " + str(acc_ridge_range))
```

2.3 Lasso regression

```python
# Calculating the r-squared score of the model
y_pred_lasso = lasso.predict(X_test)
print(y_pred_lasso[:10])
print(y_test[:10])
r2_score_lasso = r2_score(y_test, y_pred_lasso)
print("R-squared score: " + str(r2_score_lasso))

# Calculating the range accuracy score of the model. Here, a prediction is considered accurate if it is +-3 positions away from the true value
y_test_list = y_test.tolist()
acc_preds_lasso = 0
for i in range(len(y_pred_lasso)):
    if ((y_pred_lasso[i]-3 <= y_test_list[i]) and (y_test_list[i] <= y_pred_lasso[i] + 3)):
        acc_preds_lasso += 1
        
acc_lasso_range = acc_preds_lasso/len(y_test_list)
print("Range accuracy: " + str(acc_lasso_range))
```

```python
print("R-squared scores:")
print("LR: " + str(r2_score_lr), "Ridge: " + str(r2_score_ridge), "Lasso: " + str(r2_score_lasso))
print()
print("Range accuracies: ")
print("LR: " + str(acc_lr_range), "Ridge: " + str(acc_ridge_range), "Lasso: " + str(acc_lasso_range))
```

From this, we can see that the best regression model is the Lasso regression model.

Taking all of the aquired information into account, the choice for the best model would have to be between the Random Forest classifier and the Lasso regression model. Since range accuracy is higher when using the Random Forest classifier, it is probably the best choice.
