## Adding audio features

Võtab suht kaua et runnida

Kirjutasin tulemuse csv faili ka (allpool)

```python
import os
import json

directory = os.path.abspath("audio_features")

all_features = pd.DataFrame()

for root, dirs, files in os.walk(directory):
    for file in files:
        with open(root + "/" + file) as data_file:
            json_data = json.load(data_file)
        features = pd.json_normalize(json_data)
        features["year"] = int(root.split("/")[-1])
        features["to_country"] = file.split("_")[0]
        features = features[features.columns.drop(list(features.filter(regex='metadata.+')))] # removing metadata
        all_features = all_features.append(features)

# getting rid of some extracted features that are tricky to deal with
# and that we most probably don't need
for column in all_features:
    if type(all_features[column].values[0]) == list:
        all_features = all_features.drop(columns=column)

data = data.merge(all_features, how="outer")
data
```

```python
# Remove all songs that weren't available on YT and as a result don't have extracted audio features
# There are 48 such songs
data = data[data["tonal.key_scale"].notna()]
```

```python
data.to_csv("dataset_with_audio_features.csv", index=False)
```
